## Data preprocessing

### I drop the column 'car_ID'  and 'CarName' because it's irrelevant for my model

```python
car.drop('car_ID', axis=1, inplace=True)
```

```python
car.drop('CarName', axis=1, inplace=True)
```

### Splitting the dataset into independent and dependent variable

```python
X = car.drop(columns=["price"])
y = car["price"]
```

```python
encoder = LabelEncoder()

car_columns = ['fueltype', 'aspiration', 'doornumber', 'enginelocation']
for i in car_columns:
    if X[i].dtype == 'object':
        X[i] = encoder.fit_transform(X[i])
```

### OneHotEncoding the columns with object

```python
car_columns = ['symboling', 'carbody', 'drivewheel', 'enginetype', 'cylindernumber', 'fuelsystem']
columnTransformer = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), car_columns)], remainder='passthrough')
X = np.array(columnTransformer.fit_transform(X))
```

```python
type(X)
```

```python
pd.DataFrame(X)
```

### Splitting dataset into training and test set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20)
```

### Scalling 

```python
pd.DataFrame(X[: ,39:42])
```

```python
# scaler = StandardScaler()
scaler = RobustScaler()
X_train[:,40:] = scaler.fit_transform(X_train[:,40:])
X_test[:,40:] = scaler.transform(X_test[:,40:])
```

### Model Regression

```python
def build_RegressionModel(model):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    X_train_poly = poly_features.fit_transform(X_train)
    X_test_poly = poly_features.fit_transform(X_test)
    linear_regressor.fit(X_train_poly, y_train)
    y_pred = linear_regressor.predict(X_test_poly)

    print("Mean Absolute Error: ", mean_absolute_error(y_test, y_pred))
    print("r2_score: ", r2_score(y_test, y_pred))
    
    plt.figure(figsize=(6,4))
    plt.scatter(y_test, y_pred)
    plt.axis('tight')
    plt.xlabel('Price')
    plt.ylabel('Predicted price')
    plt.tight_layout()
```

```python
lr = LinearRegression()
build_RegressionModel(lr)
```

```python
rForest = RandomForestRegressor(n_estimators = 100)
build_RegressionModel(rForest)
```

```python
tree = DecisionTreeRegressor()
build_RegressionModel(tree)
```

```python
neighbors = KNeighborsRegressor(n_neighbors = 5)
build_RegressionModel(neighbors)
```

```python
svr = SVR(kernel="rbf")
build_RegressionModel(svr)
```

### Polynomial Regression 

```python
poly_features = PolynomialFeatures(degree= 2)
X_train_poly = poly_features.fit_transform(X_train)
X_test_poly = poly_features.fit_transform(X_test)
linear_regressor.fit(X_train_poly, y_train)
y_pred = linear_regressor.predict(X_test_poly)

print("Mean Absolute Error: ", mean_absolute_error(y_test, y_pred))
print("r2_score: ", r2_score(y_test, y_pred))

plt.figure(figsize=(6,4))
plt.scatter(y_test, y_pred)
plt.axis('tight')
plt.xlabel('Price')
plt.ylabel('Predicted price')
plt.tight_layout()
```

```python
models = {    
    'kNeighbors': KNeighborsRegressor(),   
    'rForest': RandomForestRegressor(),
    'svc_rbf': SVR(),   
}

gridsearch = {   
    'kNeighbors': {'n_neighbors': [1, 3, 5,10, 50]},    
    'rForest': {'n_estimators': [100, 200, 300], 'max_depth': [None, 5, 10, 15]},
    'svc_rbf': {'C': [0.1, 1, 10], 'gamma': [0.1, 0.01], 'kernel': ['rbf']},
    'boost': {'n_estimators': [20, 50, 100]}    
}

for classifierModel, model in models.items():
    if classifierModel in gridsearch:
        param_grid = gridsearch[classifierModel]
        grid_model = GridSearchCV(model, param_grid, cv=5,n_jobs=-1)
        grid_model.fit(X_train, y_train)
        best_model = grid_model.best_estimator_
        y_pred = best_model.predict(X_test)        
        
        print(classifierModel)            
        print(f'Best Parameters: {grid_model.best_params_}')        
        print("Mean Absolute Error: ", mean_absolute_error(y_test, y_pred))
        print("r2_score: ", r2_score(y_test, y_pred))
        
        plt.figure(figsize=(6,4))
        plt.scatter(y_test, y_pred)
        plt.title(classifierModel)
        plt.axis('tight')
        plt.xlabel('Price')
        plt.ylabel('Predicted price')
        plt.tight_layout()
```
