## Data preprocessing

### Drop index 42 and 43 becouse they are dublicated

```python
burger.drop([42,43],inplace=True)
```

### I drop the column 'Item' because it's irrelevant for my model

```python
burger.drop('Item', axis=1, inplace=True)
```

### Splitting the dataset into independent and dependent variable

```python
X = burger.drop(columns=['Category', 'Calories', 'Fat (g)',
       'Saturated Fat (g)', 'Trans Fat (g)', 'Cholesterol (mg)', 'Sodium (mg)',
       'Total Carb (g)', 'Dietary Fiber (g)', 'Sugars (g)', 'Weight Watchers', 'Protein (g)'])
y = burger['Cholesterol (mg)']
```

```python
X = burger.iloc[:, 2:-1].values
y = burger.iloc[:, -1].values
```

```python
X = burger.drop(columns='Cholesterol (mg)')
y = burger['Cholesterol (mg)']
```

```python
print(X)
```

```python
pd.DataFrame(X)
```

### OneHotEncoding the column 'Catagory'

```python
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
pd.DataFrame(X)
```

### Splitting dataset into training and test set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
linear_regressor = LinearRegression()
linear_regressor.fit(X_train, y_train)
```

```python
result = linear_regressor.predict(X_test)
plt.scatter(X_test, y_test)
plt.scatter(X_test, result, color="red")
plt.plot(X_test, result, color="green")
```

### Model Regression

```python
from wordcloud import WordCloud, ImageColorGenerator
text = ''' Inlämningsuppgift 1 del 1 om regression. Lets build some models on Dataset Burger King's menu.. hihi
'''
wordcloud = WordCloud().generate(text)

# Display the generator image
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
def build_RegressionModel(model):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
 
    print("Mean Absolute Error: ", mean_absolute_error(y_test, y_pred))
    print("r2_score: ", r2_score(y_test, y_pred)) 
    plt.scatter(y_test, y_pred)   
```

```python
lr = LinearRegression()
build_RegressionModel(lr)
```

```python
lr = LinearRegression()
build_RegressionModel(lr)
```

```python
rForest = RandomForestRegressor(n_estimators = 100)
build_RegressionModel(rForest)
```

```python
tree = DecisionTreeRegressor()
build_RegressionModel(tree)
```

```python
neighbors = KNeighborsRegressor(n_neighbors = 5)
build_RegressionModel(neighbors)
```

```python
svr = SVR(kernel="rbf")
build_RegressionModel(svr)
```
