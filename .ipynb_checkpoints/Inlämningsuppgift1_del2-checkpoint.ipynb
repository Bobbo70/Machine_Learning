## Data preprocessing

### I drop the column 'customerID' because it's irrelevant for my model

```python
churn.drop('customerID', axis=1, inplace=True)
```

### The column 'TotalCharges' have some missing value and should be in float not object type

```python
churn.query('TotalCharges == " "')
```

### Count the median value from the column

```python
#np.median(churn.query('TotalCharges != " "')['TotalCharges'].astype('float'))
```

### Convert the column into float and replace missing value with median

```python
churn['TotalCharges']=churn['TotalCharges'].replace(" ","0.0")
churn['TotalCharges']=churn['TotalCharges'].astype('float')
```

```python
churn
```

### Check how many unique value of each object

```python
for column in churn.columns:
    if (churn[column].dtypes == 'object'):
        print(f"{column}: {churn[column].unique()}")
```

### Replace columns containinging 'No internet service' and 'No phone service' with values No 

```python
churn.replace({'No phone service': 'No'}, inplace=True)
churn.replace({'No internet service': 'No'}, inplace=True)
```

### Encoding data

```python
encoder = LabelEncoder()

churn_columns = ['gender','Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup',
                'DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']

for i in churn_columns:
    if churn[i].dtype == 'object':
        churn[i] = encoder.fit_transform(churn[i])
```

```python
churn
```

### Splitting the dataset into independent and dependent variable

```python
X = churn.drop(columns="Churn")
y = churn["Churn"]
```

```python
pd.DataFrame(X)
```

```python
pd.DataFrame(y)
```

### OneHotEncoding 'InternetService, Contract, PaymentMethod'

```python
churn_columns = ["InternetService","Contract","PaymentMethod"]
columnTransformer = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), churn_columns)], remainder='passthrough')
X = np.array(columnTransformer.fit_transform(X))
```

### Splitting dataset into training and test set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 20)
```

```python
pd.DataFrame(X)
```

### Scalling 

```python
pd.DataFrame(X[1:10, :16]) # Just looking for the column 'tenure'
```

```python
pd.DataFrame(X)[[14,24,25]]
```

### Scaler columns 'tenure, MonthlyCharges, TotalCharges' all other are encoding

```python
#scaler = StandardScaler()
scaler = RobustScaler()
#X_train[:, 24:] = scaler.fit_transform(X_train[:,24:])
X_train[:, [14,24,25]] = scaler.fit_transform(X_train[:, [14,24,25]])
X_test[:, [14,24,25]] = scaler.transform(X_test[:, [14,24,25]])
```

```python
pd.DataFrame(X_train)
```

## Models Classification

```python
def build_classifierModel(model):    
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    cm = confusion_matrix(y_test, y_pred, labels=model.classes_)    
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
    disp.plot()   
   
    print("Testing Accuracy Score: ", accuracy_score(y_test, y_pred))    
    print("******* classification_report ********")
    print(classification_report(y_test, y_pred))
    
    print("******* Confusion Matrix ********")
    plt.show()
```

### LogisticRegression

```python
lr = LogisticRegression(random_state=0)
build_classifierModel(lr)
```

### KNeighborsClassifier

```python
kNeighbors = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
build_classifierModel(kNeighbors)
```

### DecisionTreeClassifier

```python
tree = DecisionTreeClassifier(criterion = 'entropy', random_state=0)
build_classifierModel(tree)
```

### RandomForestClassifier

```python
rForest = RandomForestClassifier(n_estimators=10, criterion = 'entropy', random_state = 0)
build_classifierModel(rForest)
```

### Support Vector Machine - RBF

```python
svc_rbf = SVC(kernel = 'rbf', random_state=0)
build_classifierModel(svc_rbf)
```

### AdaBoostClassifier

```python
boost = AdaBoostClassifier(n_estimators=50, learning_rate=1, algorithm="SAMME.R")
build_classifierModel(boost)
```

## Models classificering with other values. I use param_grid with six diffrents models

```python
models = {
    'lr' : LogisticRegression(),
    'kNeighbors': KNeighborsClassifier(),
    'tree': DecisionTreeClassifier(),
    'rForest': RandomForestClassifier(),
    'svc_rbf': SVC(),
    'boost': AdaBoostClassifier()    
}

gridsearch = {
    'lr': {'C': [0.1, 1, 10],'max_iter':[500,1000]},
    'kNeighbors': {'n_neighbors':[3, 5, 10, 20, 50, 100], 'weights':['distance','uniform']},
    'tree': {'criterion':['gini', 'entropy'], 'max_depth': [None, 3, 5, 7]},
    'rForest': {'n_estimators': [100, 200, 300], 'max_depth': [None, 5, 10, 15]},
    'svc_rbf': {'C': [0.1, 1, 10], 'gamma': [0.1, 0.01], 'kernel': ['rbf']},
    'boost': {'n_estimators': [20, 50, 100]}    
}

for classifierModel, model in models.items():
    if classifierModel in gridsearch:
        param_grid = gridsearch[classifierModel]
        grid_model = GridSearchCV(model, param_grid, cv=5,n_jobs=-1)
        grid_model.fit(X_train, y_train)
        best_model = grid_model.best_estimator_
        y_pred = best_model.predict(X_test)
         
        cm = confusion_matrix(y_test, y_pred, labels=grid_model.classes_)    
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=grid_model.classes_)
        disp.plot() 
        
        print(classifierModel)
        print("Accuracy Score: ", accuracy_score(y_test, y_pred))       
        print(f'Best Parameters: {grid_model.best_params_}')
        
        print("******* classification_report ********")
        print(classification_report(y_test, y_pred))
        
        print("******* Confusion Matrix ********")
        plt.show()
```
